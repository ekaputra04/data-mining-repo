# Compare Models

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Menggabungkan Hasil


```python
files = [
    'linear_regression_results.csv',
    'ridge_regression_results.csv',
    'random_forest_results.csv',
    'xgboost_results.csv'
]
results = []
for file in files:
    df = pd.read_csv(file)
    results.append(df)
results_df = pd.concat(results, ignore_index=True)
```

## Visualisasi Test R²


```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Model', y='Test R2', data=results_df)
plt.title('Perbandingan Test R² Score Antar Model')
plt.xlabel('Model')
plt.ylabel('Test R² Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Visualisasi Test RMSE


```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Model', y='Test RMSE', data=results_df)
plt.title('Perbandingan Test RMSE Antar Model')
plt.xlabel('Model')
plt.ylabel('Test RMSE')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Menyimpan Hasil Gabungan


```python
results_df.to_csv('all_model_results.csv', index=False)
print("\nHasil gabungan disimpan sebagai 'all_model_results.csv'")
print("\n=== Ringkasan Hasil Semua Model ===")
print(results_df[['Model', 'Test MAE', 'Test RMSE', 'Test R2', 'CV R2 Mean', 'CV R2 Std']])
```
